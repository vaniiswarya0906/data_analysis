```python
import numpy as np
#creating arrays
arr=np.array([1,2,3,4,5])
print(arr)

a=np.zeros((3,3),dtype=int)
print(a)

b=np.ones((2,2))
print(b)

arange=np.arange(10)
print(arange)

#array manipulation

#reshaping function
manip=arr.reshape(5,1)
print(manip)

#slice function
sliced=arr[2:4]
print(sliced)

#vertical stack function
a=np.array([1,2,3,4,5])
b=np.array([6,7,8,9,10])
c=np.vstack(a+b)
print(c)

#normal stack function
a=np.array([1,2,3,4,5])
b=np.array([6,7,8,9,10])
c=np.stack(a+b)
print(c)

#split function
a=np.array([1,2,3,4,5,6])
b=np.split(a,2)
print(b)

#transpose function
#2d array declaration
a=np.array([[1,2,3],[4,5,6]])
b=a.T
print(b)

```

```python
import numpy as np

#element wise operations
a1=np.array([1,2,3,4,5])
a2=np.array([2,3,4,5,6])
result=a1+a2
print(result)

#broadcasting
a1=np.array([1,2,3,4,5])
result1=a1+3
print(result1)
```

```python
#linear algebra with numpy

a1=np.array([[2,3],[3,2]])
a2=np.array([[5,4],[4,5]])
print(a1)
print(a2)
res=np.dot(a1,a2) #dot product
print(res)
d=np.linalg.eig(res )
print(d)

a1=np.array([[2,3,5],[2,3,6]])
print(a1)
b=np.sum(a1)
print(b)
a2=np.array([[5,6,7],[7,8,9]])
b2=np.sum(a1+a2)
print(b2)
a2=np.array([[5,6,7],[7,8,9]])
c=np.sum(a2, axis=0)#row wise addition

d=np.sum(a2, axis=1)#column wise addition
print(c)
print(d)

```

```python
#statistical operations

a=np.array([1,2,3,4,5])
print(a)
c=np.mean(a)#mean operation
print(c)
d=np.median(a)#median operation
print(d)
e=np.var(a)#variant operation
print(e)
f=np.std(a)#standard deviation operation
print(f)
```
