```python
import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset("tips")#load example dataset
sns.scatterplot(x="total_bill",y="tip",data=tips)#create a scatter plot
plt.title("scatter plot of total billvs.tip")
plt.xlabel("total bill")
plt.ylabel("tip")
plt.show
```

```python
tips=sns.load_dataset("diamonds")
print(tips)
sns.barplot(x="carat",y="price",data=tips)#create a bar plot
plt.title("plot of diamonds price and its carat")
plt.xlabel("carat")
plt.ylabel("price")
plt.show()
sns.violinplot(x="carat",y="price",data=tips)#create a violin plot
plt.title("plot of diamonds price and its carat")
plt.xlabel("carat")
plt.ylabel("price")
plt.show()
```

```python
iris=sns.load_dataset("iris")
print(iris)
correlation_matrix=iris.corr()
sns.heatmap(correlation_matrix,annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap of Iris dataset")
plt.show()
```
